```python
import numpy as np
```

```python
import json
with open('loc_data.json','r') as file:
    location_dict = json.load(file)
```

```python

with open('location_dict.json','r') as file:
    biggie = json.load(file)
```

```python
for each in location_dict:
    location_dict[each]['country'] = biggie[each][0]['components']['country']
```

```python
for each in location_dict:
    if location_dict[each]['country'] == "United States of America":
        location_dict[each]['state'] = biggie[each][0]['components']['state']
```
